## Lecture 2: Artificial Linearily Seperable Data

In this notebook we will generate an artificial collection of linearily seperable data points. Such collections can be useful when testing machine learning algorithms. 

For this notebook you will need to import the following packages:
* Plots [documentation](http://docs.juliaplots.org/latest/) 

```julia
using Plots
```

The idea can be summarized below:
1. Create a linear function whose graph passed through $[0,1]\times[0,1]$.
2. Generate random data points in $[0,1]\times[0,1]$.
3. Seperate these data points into two sets, one of which lies above the line and the other below.

```julia
# Generate a random linear function
coeff = rand(2)
y(x) = coeff[1]*x + coeff[2]

# Generate random tuples of data
X = rand(100, 2)
D = [(X[i,1], X[i,2]) for i = 1:100]
```

```julia
# above will store data points above the random linear function
above = []

# below will store data points below the random linear function 
below = []

for data in D
    if y(data[1]) < data[2]
        push!(above, data)
    else
        push!(below, data)
    end
end  
```

```julia
scatter([x for x in above], label = "upper data")
scatter!([x for x in below], label = "lower data")
plot!(y, label = "seperator")
```

```julia
# Condense this idea into a simple function 
function make_linearily_seperable_data(n)
    coeff = rand(n)
    X = rand(n,2)
    above = []
    below = []
    
    for i = 1:n
        if coeff[1]*X[i,1] + coeff[2] < X[i,2]
            push!(above, (X[i,1], X[i,2]))
        end
        if coeff[1]*X[i,1] + coeff[2] > X[i,2]
            push!(below, (X[i,1], X[i,2]))
        end
    end
    
    return [above, below]
end
```

```julia
D = make_linearily_seperable_data(20)
scatter([x for x in D[1]])
scatter!([x for x in D[2]])
```
