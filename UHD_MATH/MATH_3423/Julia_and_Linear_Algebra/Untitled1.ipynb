# Ordinary Differential Equations: Euler's Method

In this notebook we implement Euler's method for solving first order differential equation. 

$\textbf{Euler's Method}$
* Initial value: $(x_0, y_0)$
* Differential equation: $\frac{dy}{dx} = f(x,y)$
* Euler's update: $y(x_{n+1}) = y(x_n) + hf(x_n, y_n)$

```julia
# Differential equation
f(x,y) = 6.0 - 2.0*(x/y)

# Initial value
initial_value = (3.0, 1.0);
```

```julia
function Eulers_method(dy, initial_value, n, h)
    data = [(initial_value[1], initial_value[2])]
    x = data[1][1]
    y = data[1][2]
    
    while size(data)[1] < n
        y += y + h*dy(x, y)
        push!(data, (x,y))
        x += h*x
    end
    
    return data
end    
```

```julia
D = Eulers_method(f, initial_value, 100, .25)
```

```julia
using Plots
scatter([x for x in D])
plot!([x for x in D])
```

```julia
h(x,y) = y
D = Eulers_method(h, (0.0, 5.0), 2, .00000000005)
```

```julia
scatter([x for x in D])
plot!([x for x in D])
```

```julia
g(x,y) = (x^3)/3
D = Eulers_method(g, (0.0, 1.0), 5, .0000005)
```

```julia
scatter([x for x in D])
plot!([x for x in D])
```

```julia
using DifferentialEquations
using Plots
using ParameterizedFunctions
pyplot()
lorenz = @ode_def Lorenz begin
  dx = σ*(y-x)
  dy = ρ*x-y-x*z
  dz = x*y-β*z
end σ = 10. β = 8./3. ρ => 28.
u0 = [1., 5., 10.]
tspan = (0., 2.)
prob = ODEProblem(lorenz, u0, tspan)
sol = solve(prob)
animate(sol,vars=(:x,:y,:z),xlims=(-20,20),ylims=(-15,20),zlims=(10,40))
```
