```julia
using Plots
M = [1 1 2 3 3 1.0; 1 2 3 2 1 1.0]
```

```julia
plot([x for x in zip(M[1,:], M[2,:])])
```

```julia
plot([x for x in zip(M[1,:], M[2,:])])
A1 = [4 0; 0 4.0]
N = A1*M
plot!([x for x in zip(N[1,:], N[2,:])])
```

```julia
plot([x for x in zip(M[1,:], M[2,:])])
A1 = [2 0; 0 2]
N = A1*M
plot!([x for x in zip(N[1,:], N[2,:])])
A2 = [3 0; 0 3]
N = A2*M
plot!([x for x in zip(N[1,:], N[2,:])])
A3 = [4 0; 0 4]
N = A3*M
plot!([x for x in zip(N[1,:], N[2,:])])
A4 = [5 0; 0 5]
N = A4*M
plot!([x for x in zip(N[1,:], N[2,:])])
```
